```python
#pandas is a library,its used to analyse the data
import pandas as pd
```

```python
#datapreprocessing
dataset=pd.read_csv('50_Startups.csv')
dataset
```

```python
#getdummiesfromthedata
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
#getonlycolumns
dataset.columns
```

```python
#splitinputandoutput
#inputvariable
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
independent
```

```python
#outputvariable
dependent=dataset[['Profit']]
dependent
```

```python
#splittrainandtestdata
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
x_train,x_test,y_train,y_test
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='friedman_mse', splitter='best')#Passing the HyperTuning parameter
regressor=regressor.fit(x_train,y_train)
```

```python
#weareusingMatplotlip
import matplotlib.pyplot as plt#Genratinggraph#graphpurpose
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#prediction
y_pred=regressor.predict(x_test)
```

```python
#Evaluationmetrics
#CompareActualandPredictedvalue
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
